### Perparation ###

Use one module of the PV-Mini-Box and connect it to the load patch panel with no resistors attached to the load terminals (open circuit). 

Configure one multimeter as **Amperemeter** by turning the wheel selecting the **mA** measurement mode. Attach two 50cm long lab cables to the according sockets of the Amperemeter.

Configure the other multimeter as **Voltmeter** by turning the wheel selecting the **V** measurement mode. Attach two 50cm long lab cables to the according sockets of the Voltmeter.

**Connect the voltmeter directly (in parallel) to the output of your selected PV module,** which is already connected to the load panel.

**Mount the Amperemeter (in series) between one of the PV module outputs and the load panel.** Therefore you must open one of the connections you established before between PV module and load panel at the load panel side. Put the now free cable connector into one socket of the Amperemeter. Use an additional cable to connect the Amperemeter to the load panel.

### Ex. 3.1.1: Light source and short circuit current

**CAUTION!!!**<br>
**THE INCANDECENT LIGHT BULB IS GETTING HOT!**<br>
**THE MODULES ARE GETTING HOT!**<br>
**BE CAREFUL! DO NOT BURN YOURSELF!**<br>
**DO NOT IRRADIATE THE MODULES TOO STRINGLY!**<br>
**LIMIT THE SHORT CIRCUIT CURRENT TO 20mA!**
    
    
Create a **short circuit in the load panel** by connecting the pin headers of **load terminal A and B** with a jumper cable.

Switch on the incandecent light bulb an turn the dimmer to max. 

**Adjust the height of the light source** to generate a short circuit current of **approximately 25 mA.** 

Then **dimm the light source** to a short circuit current of **approximately 20 mA.** The current may **drift** a bit. It is good enough if you set the current **approximately** to 20 mA, i.e. **somewhere between 19 mA and 21 mA.**

Illuminate all 4 modules as **uniformly** as possible. 




**Question:** How does the short circuit current $I_\mathrm{SC}$ depend on the irradiance? Dimm the light down and observe. What happens it you switch of the light bumb completely?

**Answer:** The higher the irradiance, the higher the Isc. The Isc was 0.24 when the light bulb was switched off. It wasn't completely zero because of the surrounding light, i.e. ceiling light.

### Ex. 3.1.2: Open circuit voltage $V_\mathrm{OC}$

Dimm the light source such that $I_\mathrm{SC} \approx 20 \mathrm{mA}.$

**Question:** Which $V_\mathrm{OC}$ does the module show? How does it depent on the light intensity?

**Answer:** It is 5.90 V.When the light bulb was turned off, i.e. lowest light intensity, it was 3.38 V. There is direct proportionality between Voc and light intensity.

**Question:** Which $V_\mathrm{OC}$ woould you expect from a single Silicon (Si) PV-**Cell**? Why is the open circuit voltage you measure higher? How many times is is higher? What does that mean for the internal setup of the mini module?

**Answer:** A cell would have a Voc of magnitude of 0.5 V (Quaschning, p. 187). In our PV module 1, there are multiple cells, around 12 cells to get the Voc of around 6 V.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load data
df_single = pd.read_csv("./data/I-V_single_module.txt", comment = "#", names = ["V", "ImA", "R"])
df_single.head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,4), dpi = 136)

ax[0].plot(df_single["V"], df_single["ImA"], "*-") 

ax[0].set_title("I-V Curve of a Single PV Mini Module")
ax[0].set_xlabel("Voltage $V$ in V")
ax[0].set_ylabel("Current $I$ in mA")
ax[0].grid(True)

ax[1].plot(df_single["V"], df_single["ImA"]*df_single["V"], "*-") 

ax[1].set_title("Power Curve of a Single PV Mini Module")
ax[1].set_xlabel("Voltage $V$ in V")
ax[1].set_ylabel("Power $P$ in mW")
ax[1].grid(True)


plt.plot()
```
